**Example**: 
$$\sum_{n=0}^{\infty} \frac{(\text{i}\pi)^n}{n!}$$

```python
# Another example with an imaginary number in a series
n = sympy.symbols('n')
S = sympy.Sum((sympy.I * sympy.pi)**n / sympy.factorial(n), (n,0,sympy.oo))
S
```

```python
S.doit()
```

Part (b): Determine $p$ such that $\sum_{k=0}^{\infty} (1-p)^k = \frac{17}{5}$.

Answer to (b): The series has value 17/5 if $p = \frac{5}{17}$ since the value of the sum in general is $\frac{1}{p}$. Since $\frac{5}{17}$ satisfies $0 < \frac{5}{17} < 2$, this value leads to a convergent series.
