## Example: lateral surface of champagne glass (page 40 in lecture notes)

Need to compute the integral
$$\int_0^h 2\pi \cdot \left( ya + \frac{a^2}{4}\right)^{\frac12} \text{d}y.$$

```python
import sympy as sy
a,y,h = sy.symbols('a,y,h')
```

```python
sy.integrate(2*sy.pi* (y*a + a**2/4)**(1/2),
             (y,0,h))
```

## Exercise sheet 7, Ex. 2

**Step 1**: In order to determine the limits of the integral, we compute the points of intersection of the two functions:

```python
x = sy.symbols('x')
sy.solve(2*x**2 - sy.S(1)/5 - (-2*x**2 + sy.S(4)/5))
```

**Step 2:** Compute the integral
$$\int_{-\frac12}^{\frac12} -\left(\color{red}{2x^2 -\frac15}\right) + \left(\color{blue}{-2x^2+\frac45}\right) \text{d}x.$$

```python
sy.integrate(-(2*x**2 - sy.S(1)/5) + (-2*x**2 + sy.S(4)/5),
             (x,-sy.S(1)/2,sy.S(1)/2))
```

## Exercise sheet 7, exercise 5

**Step 1**: Visualize the two functions.
For that, we resolve $y^2-x^2=1$ with respect to $x$:
$$x(y) = \sqrt{y^2-1}$$

```python
import numpy as np
y1 = np.linspace(0,3,100)
y2 = np.linspace(1,3,100)
x1 = y1
x2 = np.sqrt(y2**2-1)
```

```python
import matplotlib.pyplot as plt
plt.plot(x1,y1)
plt.plot(x2,y2)
```

**Step 2**: Compute volume as 
$$V = \pi \int_1^3 \left(\sqrt{y^2-1}\right)^2 \text{d}y = \pi\int_1^3 \left( y^2-1\right) \text{d}y.$$

```python
y = sy.symbols('y')
V = sy.pi * sy.integrate(y**2-1,(y,1,3))
V
```
