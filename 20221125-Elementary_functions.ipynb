### Fitting an exponential function to data points of the world population

Note: The code below uses topics **not** treated in this class (specifically importing files in Python and using the `curve_fit()` function from the `scipy` package). It is for illustrative purposes. These things will **not** be required in the exam!

```python
#Import data and save to NumPy array
import os

filename = 'world_population.txt'
file = open(filename,'r')
content = file.readlines()
n = len(content)
data = np.zeros([n-1,2])

for i in range(1,n):
    x,y = content[i].split()
    x,y = float(x), float(y)
    data[i-1,0] = x
    data[i-1,1] = y

print(data)
```

```python
def exp_model(x, a, b):
    return a* (b**x)
```

```python
from scipy.optimize import curve_fit
popt, pcov = curve_fit(
    exp_model, data[:,0], data[:,1], p0 = [1650,1])
```

```python
print('Model: p(t)= {:.2f} * {:.5f}^t'.format(
    popt[0], popt[1]))
```

```python
plt.scatter(data[:,0], data[:,1])
t = np.linspace(0,110,100)
p = exp_model(t, popt[0], popt[1])
plt.plot(t,p)
```
