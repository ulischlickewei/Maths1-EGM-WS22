## 2) Limits of functions with SymPy

```python
import sympy as sy
x = sy.symbols('x')
```

We can use the `sy.limit` function for computing the limit of a function or more generally of any symbolic expression. For that sake we pass three arguments to `limit()`:
- The algebraic expression / function under investigation (in our case: `sy.sin(x)/x`
- The variable for which we consider the limiting behaviour (in our case: `x`)
- The value to where the variable above should tend (in our case: `0`)

NOTE: In the symbolic expression we need to use the Sympy `sin()` function, not the NumPy `sin()`.

```python
# Example 1 from above
sy.limit(sy.sin(x)/x, x, 0)
```

```python
# Example 2 from above
sy.limit(sy.sin(sy.pi/x), x, 0)
```

```python
# Example 3: lim of 1/x for x to 0
sy.limit(1/x, x, 0)
```
